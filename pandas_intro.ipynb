### Modifying the Data

```python
d = {'name': ['Wally']}
tmp_df = pd.DataFrame(data=d)
tmp_df
```

```python
d = {'name': ['Wally']}
tmp_df = pd.DataFrame(data=d)
roster = pd.concat([roster, tmp_df], ignore_index=True)
roster
```
